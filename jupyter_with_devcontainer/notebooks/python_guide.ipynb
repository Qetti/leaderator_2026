### 1.2 როგორ მუშაობს Python — კომპილაცია vs ინტერპრეტაცია

პროგრამირების ენებს ორ კატეგორიად ყოფენ იმის მიხედვით, როგორ "თარგმნიან" ადამიანის მიერ დაწერილ კოდს კომპიუტერის ენაზე.

**კომპილირებადი ენები (C, C++, Rust, Go):**
კომპილატორი მთელ კოდს ერთიანად თარგმნის მანქანურ ენაზე (ნულები და ერთიანები). შედეგი — ბინარული ფაილი (.exe), რომელსაც პროცესორი პირდაპირ ასრულებს. სწრაფია, მაგრამ კომპილაცია დროს მოითხოვს.

```
source code (.c) → [compiler] → machine code (.exe) → CPU ასრულებს
```

**ინტერპრეტირებადი ენები (ძველი BASIC):**
ინტერპრეტატორი კოდს ხაზ-ხაზ კითხულობს და მაშინვე ასრულებს. ნელია, მაგრამ მოსახერხებელია — კომპილაციას არ ელოდებით.

**Python — ჰიბრიდია:**

```
source code (.py)
    ↓  [ეტაპი 1: კომპილაცია bytecode-ად]
bytecode (.pyc ფაილები __pycache__/ ფოლდერში)
    ↓  [ეტაპი 2: Python Virtual Machine ინტერპრეტაცია]
შედეგი
```

1. **ეტაპი 1:** CPython კომპილატორი `.py` ფაილს **bytecode**-ად გარდაქმნის — ეს შუალედური ფორმატია, არა მანქანური კოდი. Bytecode `.pyc` ფაილებში ინახება (`__pycache__/` ფოლდერი).
2. **ეტაპი 2:** Python Virtual Machine (PVM) bytecode-ს ინტერპრეტაციით ასრულებს — ყოველ ინსტრუქციაზე "ფიქრობს" რა გააკეთოს.

სწორედ ეს მეორე ეტაპი ანელებს Python-ს. მაგრამ პრაქტიკულად, როცა NumPy-ით მატრიცების გამრავლებას ვაკეთებთ, Python მხოლოდ "მართავს" — C-ზე დაწერილ ოპტიმიზებულ კოდს იძახებს. ამიტომ: **"Python steers, C drives."**

### 2.3 მონაცემთა ტიპები (Data Types)

Python-ში ყველაფერი **ობიექტია**. ყველა მნიშვნელობას აქვს ტიპი. `type()` ფუნქციით ტიპის შემოწმება შეგიძლიათ.

#### ძირითადი ტიპები

````python
# int — მთელი რიცხვი (integer)
age = 25
big_number = 1_000_000  # _ ვიზუალური გამყოფია, Python უგულებელყოფს
negative = -42
print(type(age))  # <class 'int'>
````

````python
# float — ათწილადი (floating-point)
pi = 3.14159
temperature = -17.5
scientific = 2.5e10  # 2.5 × 10^10 = 25000000000.0
print(type(pi))  # <class 'float'>
````

````python
# str — ტექსტი (string)
name = "გიორგი"
greeting = 'გამარჯობა'  # ორმაგი და ერთმაგი ბრჭყალები ექვივალენტურია
multiline = """ეს არის
მრავალხაზიანი
ტექსტი"""
print(type(name))  # <class 'str'>
````

````python
# bool — ლოგიკური (boolean) — მხოლოდ True ან False
is_active = True
is_deleted = False
print(type(is_active))  # <class 'bool'>
````

````python
# NoneType — "არაფერი" (null სხვა ენებში)
result = None
print(type(result))  # <class 'NoneType'>
````

#### ტიპის კონვერტაცია (Type Casting)

ერთი ტიპის მნიშვნელობა შეგვიძლია სხვა ტიპად გარდავქმნათ.

````python
# str → int
age_str = "25"
age_int = int(age_str)
print(age_int + 1)  # 26
````

````python
# int → str
number = 42
text = str(number)
print("პასუხი: " + text)  # "პასუხი: 42"
# print(number + "ვაუ")
````

````python
# str → float
price_str = "19.99"
price = float(price_str)
print(price * 2)  # 39.98
````

````python
# float → int (ათწილადი ნაწილი იკარგება)
pi = 3.99
whole = int(pi)
print(whole)  # 3 (არა 4 — არ ამრგვალებს, ჭრის)
````

````python
# bool კონვერტაცია
print(bool(0))       # False (0 არის False)
print(bool(1))       # True  (ნებისმიერი არა-ნულოვანი რიცხვი არის True)
print(bool(""))      # False (ცარიელი string არის False)
print(bool("hello")) # True  (არა-ცარიელი string არის True)
print(bool([]))      # False (ცარიელი list არის False)
print(bool([1, 2]))  # True  (არა-ცარიელი list არის True)
print(bool(None))    # False (None ყოველთვის False)
````

### 2.5 ოპერატორები (Operators)

#### ინდექსაცია და Slicing

String-ში ყოველ სიმბოლოს ინდექსი აქვს. **ინდექსი 0-დან იწყება!**

```
 სიმბოლო:  P   y   t   h   o   n
 ინდექსი:  0   1   2   3   4   5
 უარყოფითი: -6  -5  -4  -3  -2  -1
```

````python
text = "Python"
````

````python
# ინდექსით წვდომა
print(text[0])    # P   (პირველი სიმბოლო)
print(text[5])    # n   (მეექვსე)
print(text[-1])   # n   (ბოლო)
print(text[-2])   # o   (ბოლოდან მეორე)
````

````python
# Slicing — [start:stop:step]
# start — საიდან (ჩათვლით)
# stop — სადამდე (გარდა!)
# step — ნაბიჯი
print(text[0:3])   # Pyt  (ინდექსი 0, 1, 2 — 3 არ შედის!)
print(text[2:5])   # tho
print(text[:3])    # Pyt  (start გამოტოვება = თავიდან)
print(text[3:])    # hon  (stop გამოტოვება = ბოლომდე)
print(text[:])     # Python (ასლი)
print(text[::2])   # Pto  (ყოველი მეორე სიმბოლო)
print(text[::-1])  # nohtyP (უკუ — string-ის "გადაბრუნება"!)
````

### 3.4 Set (set) — სიმრავლე

Set არის **კოლექცია უნიკალური ელემენტებით**. დუბლიკატები ავტომატურად იშლება.

````python
# შექმნა
fruits = {"ვაშლი", "მსხალი", "ბანანი"}
numbers = {1, 2, 3, 3, 2, 1}
print(numbers)  # {1, 2, 3} — დუბლიკატები გაქრა!
````

````python
# ცარიელი set (ფრჩხილებით dict-ია!)
empty_set = set()   # სწორი
# empty_set = {}    # ეს dict-ია!

# ელემენტის დამატება/წაშლა
fruits.add("ყურძენი")
fruits.remove("ბანანი")    # KeyError თუ არ არსებობს
fruits.discard("კივი")     # შეცდომა არ იქნება, თუ არ არსებობს
````

````python
# set-ის მათემატიკური ოპერაციები
a = {1, 2, 3, 4, 5}
b = {4, 5, 6, 7, 8}

print(a | b)     # {1, 2, 3, 4, 5, 6, 7, 8}  — გაერთიანება (union)
print(a & b)     # {4, 5}                      — თანაკვეთა (intersection)
print(a - b)     # {1, 2, 3}                   — სხვაობა (difference)
print(a ^ b)     # {1, 2, 3, 6, 7, 8}          — სიმეტრიული სხვაობა (გაერთიანებას-თანაკვეთა)
````

### 3.5 Dictionary (dict) — ლექსიკონი / ასოციაციური მასივი

Dictionary არის **key-value წყვილების** კოლექცია. ყოველ key-ს (გასაღებს) თავისი value (მნიშვნელობა) აქვს. key-ები უნდა იყოს უნიკალური და immutable (string, int, tuple).

````python
# შექმნა
person = {
    "name": "გიორგი",
    "age": 30,
    "city": "თბილისი",
    "hobbies": ["კითხვა", "ფეხბურთი"]
}
````

````python
# მნიშვნელოვანი მეთოდები
print(person.keys())    # dict_keys(['name', 'age', 'hobbies'])
print(person.values())  # dict_values(['გიორგი', 31, [...]])
print(person.items())   # dict_items([('name', 'გიორგი'), ('age', 31), ...])
````

````python
# შემოწმება
print("name" in person)      # True (key-ში ეძებს)
print("გიორგი" in person)    # False (value-ში არ ეძებს!)
````

### 4.1 პირობითი ოპერატორები — if / elif / else

პირობითი ოპერატორი კოდის სხვადასხვა ბლოკს ასრულებს პირობის მიხედვით.

````python
age = 20

if age < 13:
    print("ბავშვი")
elif age < 18:
    print("მოზარდი")
elif age < 65:
    print("ზრდასრული")
else:
    print("პენსიონერი")
# ზრდასრული
````

### 4.2 for ციკლი

`for` ციკლი iterable-ის (გამეორებადი ობიექტის) ყოველ ელემენტზე "გადის".

````python
# list-ზე იტერაცია
fruits = ["ვაშლი", "მსხალი", "ბანანი"]
for fruit in fruits:
    print(fruit)
# ვაშლი
# მსხალი
# ბანანი

# string-ზე იტერაცია (ყოველ სიმბოლოზე)
for char in "Hello":
    print(char)
# H e l l o (ყოველი ცალკე ხაზზე)

# range() — რიცხვითი დიაპაზონი
for i in range(5):       # 0, 1, 2, 3, 4
    print(i)

for i in range(2, 8):    # 2, 3, 4, 5, 6, 7
    print(i)

for i in range(0, 10, 2):  # 0, 2, 4, 6, 8 (ნაბიჯი 2)
    print(i)

for i in range(10, 0, -1):  # 10, 9, 8, ..., 1 (უკუ)
    print(i)
````

````python
# enumerate() — ინდექსი + მნიშვნელობა
fruits = ["ვაშლი", "მსხალი", "ბანანი"]
for index, fruit in enumerate(fruits):
    print(f"{index}: {fruit}")
# 0: ვაშლი
# 1: მსხალი
# 2: ბანანი

# ინდექსის საწყისი მნიშვნელობა:
for i, fruit in enumerate(fruits, start=1):
    print(f"{i}. {fruit}")
# 1. ვაშლი
# 2. მსხალი
# 3. ბანანი

# zip() — ორი list-ის პარალელური იტერაცია
names = ["ანა", "გიორგი", "მარიამი"]
ages = [25, 30, 28]
for name, age in zip(names, ages):
    print(f"{name} — {age} წლის")
# ანა — 25 წლის
# გიორგი — 30 წლის
# მარიამი — 28 წლის

# dict-ზე იტერაცია
person = {"name": "ანა", "age": 25, "city": "თბილისი"}
for key, value in person.items():
    print(f"{key}: {value}")
````

````python
# break — ციკლიდან სრული გამოსვლა
for i in range(10):
    if i == 5:
        break
    print(i)
# 0, 1, 2, 3, 4 (5-ზე გაჩერდა)

# continue — ამ იტერაციის გამოტოვება, შემდეგზე გადასვლა
for i in range(10):
    if i % 2 == 0:
        continue   # ლუწებს გამოტოვებს
    print(i)
# 1, 3, 5, 7, 9

# else ციკლში — მხოლოდ მაშინ ეშვება, თუ break არ მოხდა!
for i in range(10):
    if i == 100:   # ვერ იპოვის
        break
else:
    print("ციკლი ბოლომდე გაიარა, break არ მოხდა")
# "ციკლი ბოლომდე გაიარა, break არ მოხდა"

# პრაქტიკული მაგალითი — ელემენტის ძებნა:
numbers = [2, 4, 6, 8, 10]
target = 7
for num in numbers:
    if num == target:
        print(f"ნაპოვნია: {num}")
        break
else:
    print(f"{target} ვერ მოიძებნა")
# "7 ვერ მოიძებნა"
````

### 5.1 რა არის ფუნქცია

ფუნქცია — ეს არის კოდის ბლოკი, რომელიც კონკრეტულ ამოცანას ასრულებს და შეგვიძლიათ სახელით გამოვიძახოთ. ფუნქციები არსებობს კოდის **ორგანიზაციისთვის, გამეორების თავიდან ასაცილებლად (DRY — Don't Repeat Yourself) და წაკითხვადობისთვის**.

````python
# ფუნქციის განსაზღვრა (definition)
def greet(name):
    """ბეჭდავს მისალმებას."""
    print(f"გამარჯობა, {name}!")
````

````python
# ფუნქციის გამოძახება (call)
greet("ანა")       # გამარჯობა, ანა!
greet("გიორგი")    # გამარჯობა, გიორგი!
````

**ანატომია:**
- `def` — keyword, ფუნქციის დასაწყისი
- `greet` — ფუნქციის სახელი (snake_case)
- `(name)` — პარამეტრი (input)
- `:` — ორწერტილი
- `"""..."""` — docstring (დოკუმენტაცია)
- ბლოკი — 4 space-ით შეწეული

#### *args და **kwargs — ცვალებადი რაოდენობის არგუმენტები

````python
# *args — ნებისმიერი რაოდენობის პოზიციური არგუმენტი (tuple-ად)
def sum_all(*args):
    print(type(args))  # <class 'tuple'>
    return sum(args)

print(sum_all(1, 2, 3))      # 6
print(sum_all(1, 2, 3, 4, 5)) # 15
````

````python
# **kwargs — ნებისმიერი რაოდენობის keyword არგუმენტი (dict-ად)
def print_info(**kwargs):
    print(type(kwargs))  # <class 'dict'>
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="ანა", age=25, city="თბილისი")
# name: ანა
# age: 25
# city: თბილისი

# კომბინაცია:
def func(a, b, *args, **kwargs):
    print(f"a={a}, b={b}")
    print(f"args={args}")
    print(f"kwargs={kwargs}")

func(1, 2, 3, 4, x=5, y=6)
# a=1, b=2
# args=(3, 4)
# kwargs={'x': 5, 'y': 6}
````

### 5.4 Scope — ცვლადების ხილვადობა

````python
x = "გლობალური"  # global scope

def my_func():
    y = "ლოკალური"  # local scope — მხოლოდ ფუნქციის შიგნით
    print(x)  # გლობალურ ცვლადს ხედავს
    print(y)  # ლოკალურ ცვლადს ხედავს

my_func()
print(x)  # OK
# print(y)  # NameError! y ფუნქციის გარეთ არ არსებობს

# global keyword — გლობალური ცვლადის ცვლილება ფუნქციაში
counter = 0

def increment():
    global counter
    counter += 1

increment()
increment()
print(counter)  # 2 (global-ის გარეშე UnboundLocalError იქნებოდა)
````

````python
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n - 1)

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
````

````python
from math_helpers import factorial, is_prime

print(factorial(5))   # 120
print(is_prime(17))   # True
````

## თავი VI: შეცდომების დამუშავება და ფაილებთან მუშაობა

### 6.1 Exceptions — შეცდომების დამუშავება

პროგრამის გაშვებისას შეცდომები ხდება. Python-ში შეცდომებს **Exceptions** ეწოდება. `try/except` ბლოკით შეცდომებს "ჭერით" და მოგრაგნით.

````python
# try/except ბაზა
try:
    result = 10 / 0
except ZeroDivisionError:
    print("ნულზე გაყოფა შეუძლებელია!")
````

````python
# რამდენიმე exception
try:
    number = int(input("შეიყვანეთ რიცხვი: "))
    result = 100 / number
    print(f"შედეგი: {result}")
except ValueError:
    print("ეს რიცხვი არ არის!")
except ZeroDivisionError:
    print("ნულზე ვერ გავყოფთ!")
except Exception as e:
    print(f"უცნობი შეცდომა: {e}")
````

````python
# try/except/else/finally
try:
    file = open("data.txt")
    content = file.read()
except FileNotFoundError:
    print("ფაილი ვერ მოიძებნა")
else:
    print(f"ფაილი წარმატებით წაიკითხა: {len(content)} სიმბოლო")
finally:
    print("ეს ყოველთვის შესრულდება")
    # finally — cleanup კოდისთვის (ფაილის დახურვა, DB კავშირის გათიშვა)
````

**ხშირი Exception ტიპები:**

| Exception | მიზეზი |

| `SyntaxError` | კოდის სინტაქსი არასწორია |

| `TypeError` | არასწორი ტიპი (მაგ. `"hello" + 5`) |
| `ValueError` | სწორი ტიპი, არასწორი მნიშვნელობა (მაგ. `int("abc")`) |
| `IndexError` | list-ის ინდექსი საზღვრებს გარეთ |
| `KeyError` | dict-ში key არ არსებობს |
| `FileNotFoundError` | ფაილი ვერ მოიძებნა |
| `ZeroDivisionError` | ნულზე გაყოფა |
| `ImportError` | მოდულის import ვერ მოხერხდა |
| `AttributeError` | ობიექტს ეს ატრიბუტი/მეთოდი არ აქვს |

### 6.2 ფაილებთან მუშაობა

````python
# ფაილის წაკითხვა — with statement (context manager)
with open("data.txt", "r", encoding="utf-8") as file:
    content = file.read()
    print(content)
# with — ავტომატურად ხურავს ფაილს, შეცდომის შემთხვევაშიც!

# ხაზ-ხაზ წაკითხვა
with open("data.txt", "r", encoding="utf-8") as file:
    for line in file:
        print(line.strip())  # strip() ხაზის ბოლოს \n-ს შლის
````

````python
# ფაილში ჩაწერა
with open("output.txt", "w", encoding="utf-8") as file:
    file.write("პირველი ხაზი\n")
    file.write("მეორე ხაზი\n")
# "w" — წერა (ფაილს გადაწერს!)
# "a" — append (ფაილის ბოლოს დაამატებს)
# "r" — წაკითხვა (ნაგულისხმევი)

# JSON ფაილთან მუშაობა
import json
````

````python
# ჩაწერა
data = {"students": [{"name": "ანა", "grade": 95}]}
with open("data.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)
````

````python
# წაკითხვა
with open("data.json", "r", encoding="utf-8") as f:
    loaded = json.load(f)
    print(loaded["students"][0]["name"])  # ანა
````

**`with` statement-ის მნიშვნელობა:** `with` არის **context manager** — ის გარანტირებულად ხურავს ფაილს, მაშინაც კი, თუ შეცდომა მოხდება. `with`-ის გარეშე `file.close()` ხელით უნდა გამოიძახოთ, რაც ხშირად ავიწყდება.

## თავი VII: Package და გარემოს მართვა

### 7.1 პრობლემა — რატომ გვჭირდება გარემოს მართვა

წარმოიდგინეთ: თქვენს კომპიუტერზე ორი პროექტი გაქვთ:
- **პროექტი A** — სჭირდება `pandas 1.5` და `numpy 1.24`
- **პროექტი B** — სჭირდება `pandas 2.1` და `numpy 1.26`

თუ ორივეს სისტემის Python-ზე დააინსტალირებთ — გამოიწვევს კონფლიქტს. პანდასი ვერ გექნებათ 1.5 და 2.1 ერთდროულად.

ამას **dependency hell** ეწოდება და გადაჭრის გზა — **ვირტუალური გარემოები**.

### 7.2 ვირტუალური გარემო (Virtual Environment)

ვირტუალური გარემო — ეს არის იზოლირებული Python ინსტალაცია, სადაც პაკეტები სხვა პროექტებისგან დამოუკიდებლად ინსტალირდება.

### 7.3 ინსტრუმენტების შედარება

| ინსტრუმენტი | რა არის | სიჩქარე | როდის გამოიყენოთ |
|---|---|---|---|
| **pip + venv** | Python-ის სტანდარტი | ნელი | მარტივი პროექტი, legacy |
| **conda** | Binary package + env manager | საშუალო | ML, CUDA, non-Python deps |
| **uv** | Rust-ზე დაწერილი მოდერნული ინსტრუმენტი | ძალიან სწრაფი | ახალი პროექტები (რეკომენდებული) |

#### pip + venv — კლასიკური მიდგომა

```bash
# ვირტუალური გარემოს შექმნა
python -m venv .venv

# აქტივაცია
source .venv/bin/activate        # Linux/macOS
# .venv\Scripts\activate         # Windows

# პაკეტის ინსტალაცია
pip install pandas numpy matplotlib

# დაინსტალირებული პაკეტების სიის შენახვა
pip freeze > requirements.txt

# სხვის requirements-ის ინსტალაცია
pip install -r requirements.txt

# გამოსვლა
deactivate
```

#### conda — Data Science-ისთვის

```bash
# გარემოს შექმნა
conda create -n myenv python=3.12 pandas numpy

# აქტივაცია
conda activate myenv

# პაკეტის ინსტალაცია
conda install pytorch torchvision -c pytorch

# გამოსვლა
conda deactivate
```

conda-ს **უპირატესობა:** მართავს არა მხოლოდ Python-ის, არამედ C/C++ ბიბლიოთეკებსაც (CUDA, MKL, OpenBLAS). 

#### uv — რეკომენდებული ინსტრუმენტი

```bash
# პროექტის შექმნა
uv venv && uv pip install -r requirements.txt

# პაკეტის დამატება
uv add pandas numpy matplotlib

# Dev dependency-ის დამატება
uv add --dev ruff pytest

# Lock ფაილის მიხედვით სინქრონიზაცია
uv sync

# სკრიპტის გაშვება
uv run python script.py
```

**uv-ის უპირატესობა:** Rust-ზეა დაწერილი, 10-100x უფრო სწრაფია pip-ზე. `uv.lock` ფაილი — გარანტიას იძლევა, რომ ყველა მანქანაზე ზუსტად ერთი და იგივე ვერსიები დაყენდება.

### 7.4 requirements.txt vs pyproject.toml

```
# requirements.txt — მარტივი, პაკეტების სია
pandas==2.1.0
numpy>=1.24.0
matplotlib~=3.8.0
```

ვერსიის ოპერატორები:
- `==2.1.0` — ზუსტად ეს ვერსია
- `>=2.1.0` — ეს ან უფრო ახალი
- `~=3.8.0` — 3.8.x (minor update-ები OK, major — არა)
- `<=2.5.0` — ეს ან უფრო ძველი

```toml
# pyproject.toml — თანამედროვე, ყველაფერი ერთ ფაილში
[project]
name = "my-project"
version = "0.1.0"
requires-python = ">=3.10"
dependencies = [
    "pandas>=2.0",
    "numpy>=1.24",
]

[project.optional-dependencies]
dev = ["ruff", "pytest", "mypy"]

[tool.ruff]
line-length = 88
```

### 7.5 Lockfile — რეპროდუცირებადობა

**requirements.txt** ≠ lockfile. `pandas>=2.0` შეიძლება 2.0.0 დააინსტალიროს დღეს და 2.3.0 ხვალ. 

**Lockfile** (`uv.lock`, `poetry.lock`) ფიქსირებულ ვერსიებს ინახავს ყველა dependency-სთვის (მათ შორის sub-dependencies). შედეგი: **ნებისმიერ კომპიუტერზე, ნებისმიერ დროს — ზუსტად ერთი და იგივე გარემო.**

```bash
# uv-ით:
uv lock    # uv.lock ფაილის შექმნა/განახლება
uv sync    # uv.lock-ის მიხედვით ინსტალაცია
```
